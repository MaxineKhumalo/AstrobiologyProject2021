```python
import pickle
import numpy as np
import matplotlib.pyplot as plt
```

# Generate plots

## Sychronous plots

Open pickles

```python
new = []
o = [0,1,10,23.441,45]
for i in o:
    infile = open("obliq={}_Surface_Temperature.txt".format(str(i)),'rb')
    new.append(pickle.load(infile))
    infile.close()
#[lon,lat,ts]
```

Plot graphs

```python
i = 0
j = o[i]
[lon,lat,ts] = new[i]
im0=plt.pcolormesh(lon,lat,ts,cmap="RdBu_r",vmin=273.15-60.0,vmax=273.15+60.0,shading="Gouraud")
plt.contour(lon,lat,ts,[273.15,],colors=['gray',])
plt.colorbar(im0,label="Surface Temperature [K]")
plt.xlabel("Longitude [deg]")
plt.ylabel("Latitude [deg]")
plt.title("Obliq={} and synchronous orbit Surface Temperature".format(str(j)))
plt.savefig("obliq={}.png".format(str(j)),dpi=300)
```

Summary plots

```python
mean = np.array([])
norm = np.array([])
maximum = np.array([])
minimum = np.array([])
for i in range(len(o)):
    mean = np.append(mean,np.mean(new[i][2]))
    norm = np.append(norm,np.linalg.norm(new[i][2]))
    maximum = np.append(maximum,np.max(new[i][2]))
    minimum = np.append(minimum,np.min(new[i][2]))
```

```python
plt.plot(o,mean,'#ac956d', marker='*')
plt.xlabel("Obliquity (degrees)")
plt.ylabel("Mean surface temperature (K)")
plt.title("Synchronous orbit: Obliquity vs Surface Temperature (K)")
plt.savefig("Mean.png",dpi=300)
```

```python
plt.loglog(o,mean,'#ac956d', marker='o')
plt.xlabel("Obliquity (degrees)")
plt.ylabel("Mean surface temperature (K)")
plt.title("Log-Log Earthlike planet Obliquity vs Surface Temperature (K)")
plt.savefig("Mean LogLog.png",dpi=300)
```

```python
fig, ax = plt.subplots()
ax.plot(o,maximum,'black', marker='*')
ax.plot(o,mean,'#ac956d', marker='*')
ax.plot(o,minimum,'#761013', marker='*')
plt.legend(['max','mean','min'])
plt.xlabel("Obliquity (degrees)")
plt.ylabel("Percentage Habitable")
plt.title("Synchronous orbit: Obliquity vs Percentage Habitable")
plt.savefig("meanminmax.png",dpi=300)
```

Habitable area

```python
percentage_habitable_mean = []
for i in new:
    total = 0
    count = 0
    for j in i[2]:
        for k in j:
            total += 1
            if k>=273.15:
                count += 1
    percentage_habitable_mean.append(count/total)
            
percentage_habitable_mean     
```

Habitable plots

```python
fig, ax = plt.subplots()
ax.plot(o,percentage_habitable_mean,'#761013', marker='*')
#plt.legend(['>273.15K','mean','last','min'])
plt.xlabel("Obliquity (degrees)")
plt.ylabel("Percentage Habitable")
plt.title("Earthlike planet Obliquity vs Percentage Habitable")
plt.savefig("percentage_habitable.png",dpi=300)
```
